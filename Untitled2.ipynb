```python
def get(tickers, startdate, enddate):
    #definition
    def data(tickers):
        return web.DataReader(tickers, 'yahoo', start = startdate, end = enddate)
    stock = map(data, tickers)
    return pd.concat(stock, keys = tickers , names = ['Ticker', 'Date'])

stockdata2 = get(tickers, startdate, enddate)
```
