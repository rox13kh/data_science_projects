```python
islong = []
```

```python
for length in movies['duration']:
    if length >= 120:
        islong.append('Long')
    else:
        islong.append('Short')
```

```python
islong
```

```python
islong = pd.DataFrame(islong, columns = ['Is Long'])
```

```python
islong
```

```python
movies = movies.join(islong)
```

```python
movies
```
